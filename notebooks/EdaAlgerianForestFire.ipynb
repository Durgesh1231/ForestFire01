```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv("Algerian_forest_fires_dataset_UPDATE (2).csv" ,header = 1)
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
## DAta cleaning 
```

```python
## missing value

dataset[dataset.isnull().any(axis = 1)]
```

```python
### iss dataset main 2 region hai. tho hum 2 region bana denge 1 region ko 0 se indicate karege 2 region ko 1 se taki unn don region assani se pehchaan kar sake

dataset.loc[:122,"Region"] =0
dataset.loc[122: , 'Region'] =1
```

```python
df = dataset
```

```python
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
df[["Region"]] = df[["Region"]].astype(int)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
## removing the null values

df = df.dropna().reset_index(drop = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]   ## to accessing the any row like 122 column
```

```python
df= df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
### fix space in columns name

df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
df.info()
```

```python
### change the required columns into integer
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
### changing the other columns into float datatype
```

```python
objects = [features for features in df.columns if df[features].dtypes =='O']    ### to seprate the columns those in the form of objects
```

```python
objects
```

```python
### in the column of classes all data are present in categorical type of data the we leave the columns of classes


for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```
